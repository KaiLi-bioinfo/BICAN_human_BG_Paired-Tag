```python
import os, uuid
import snapatac2 as snap

# ============ 0) Limit threads (must set BEFORE heavy computation) ============
os.environ["OMP_NUM_THREADS"] = "4"
os.environ["OPENBLAS_NUM_THREADS"] = "4"
os.environ["MKL_NUM_THREADS"] = "4"
os.environ["VECLIB_MAXIMUM_THREADS"] = "4"
os.environ["NUMEXPR_NUM_THREADS"] = "4"

# If using Apple Silicon (M-series), this often improves stability
os.environ["RAYON_NUM_THREADS"] = "4"

# ============ 1) tempdir: use local HOME directory ============
BASE_TMP = os.path.expanduser("~/snap_tmp2")
RUN_TMP  = os.path.join(BASE_TMP, f"run_{uuid.uuid4().hex[:8]}")
os.makedirs(RUN_TMP, exist_ok=True)
print("[tempdir]", RUN_TMP)

# ============ 2) Import fragments ============
# IMPORTANT:
# Explicitly pass the number of parallel workers to the function.
# Different snapatac2 versions may use different parameter names (n_jobs / n_proc).
# The try/except block automatically handles both cases.

kwargs = dict(
    chrom_sizes=snap.genome.hg38,
    min_num_fragments=500,
    sorted_by_barcode=False,
    tempdir=RUN_TMP,
)

try:
    adatas = snap.pp.import_fragments(
        [fl for _, fl in files],
        file=[name + ".h5ad" for name, _ in files],
        n_jobs=8,          # Most common versions use n_jobs
        **kwargs
    )
except TypeError:
    adatas = snap.pp.import_fragments(
        [fl for _, fl in files],
        file=[name + ".h5ad" for name, _ in files],
        n_proc=8,          # Some older versions use n_proc instead
        **kwargs
    )

# ============ 3) Downstream processing test 25K bin & 20K features as demo ============
snap.metrics.tsse(adatas, snap.genome.hg38)
snap.pp.add_tile_matrix(adatas, bin_size=25000)
snap.pp.select_features(adatas, n_features=20000)
snap.pp.scrublet(adatas)
snap.pp.filter_doublets(adatas)
```

```python
data = snap.AnnDataSet(
    adatas =[(name, adata) for (name, _), adata in zip(files, adatas)],
    filename="BGC_BICAN_MiniAlas_H3K9me3_SnapATAC2_20260130.h5ads"
)
data
```

```python
import pandas as pd

# === 1️⃣ Split RNA barcode and sample ID from index ===
# index format example: AAACAGCCAAACAACA-1_sample_1  (barcode_sample)
idx_series = flitered_cell.index.to_series()

tmp = idx_series.str.split("_", n=1, expand=True)
flitered_cell["barcode"] = tmp[0]   # e.g. AAACAGCCAAACAACA-1
flitered_cell["sample"]  = tmp[1]   # e.g. sample_1

# Remove "-1" suffix to match the format used in translate['rna']
flitered_cell["barcode_clean"] = (
    flitered_cell["barcode"]
    .str.replace("-1$", "", regex=True)
)

# === 2️⃣ Load translation table ===
translate = pd.read_table(
    "/Volumes/Kai_T7_1/BGC_test_raw/arc_bc-translation.txt"
)

# Example format:
# atac               rna               name
# ACAGCGGGTGTGTTAC   AAACAGCCAAACAACA  arc_pair_1


# === 3️⃣ Match RNA barcodes (after removing '-1') ===
matched_translate = translate[
    translate["rna"].isin(flitered_cell["barcode_clean"])
].copy()


# === 4️⃣ Add sample information to matched entries ===
matched_translate = matched_translate.merge(
    flitered_cell[["barcode_clean", "sample"]],
    left_on="rna",
    right_on="barcode_clean",
    how="left",
)


# === 5️⃣ Construct ATAC cell IDs ===
# Format consistent with mergeed_data.obs_names
# Example: ACAGCGGGTGTGTTAC-1_sample_1
matched_translate["atac_cell"] = (
    matched_translate["atac"] + "-1_" + matched_translate["sample"]
)


# === 6️⃣ Subset ATAC AnnData using matched cells ===
matched_atac_cells = matched_translate["atac_cell"].values

subset_mergeed = mergeed_data[
    mergeed_data.obs_names.isin(matched_atac_cells)
].copy()


print(
    f"✅ Successfully extracted {subset_mergeed.n_obs} ATAC cells "
    f"corresponding to RNA cells in flitered_cell"
)
```

```python
# 1. Load ATAC → RNA mapping metadata
mapping = pd.read_csv(
    "/Users/likai/Downloads/BICAN_MiniAtlas_BGC_sample_donor_target_region_strict_MapMyCell_output_0708_V3.csv",
    comment="#"
)
mapping.head()

# Create a dictionary for ATAC barcode → RNA barcode conversion
atac_to_rna = dict(zip(translate["atac"], translate["rna"]))


# 2. Split ATAC cell_id into barcode and sample
# Format example: ACAGCGGGTGTGTTAC-1_YZ_1
merged_obs = subset_mergeed.obs.copy()
split_df = merged_obs.index.to_series().str.rpartition("_")

merged_obs["atac_barcode"] = split_df[0]
merged_obs["sample"] = split_df[2]


# 3. Remove "-1" suffix and map to corresponding RNA barcode
merged_obs["atac_core"] = merged_obs["atac_barcode"].str.replace("-1", "", regex=False)
merged_obs["rna_barcode"] = merged_obs["atac_core"].map(atac_to_rna)


# 4. Construct RNA-style cell_id
# Format: rna_barcode + "_" + sample
merged_obs["cell_id"] = merged_obs["rna_barcode"] + "_" + merged_obs["sample"]


# 5. Normalize mapping cell_id format by removing "-1"
# This ensures consistent matching between ATAC and RNA IDs
mapping_fixed = mapping.copy()
mapping_fixed["cell_id"] = mapping_fixed["cell_id"].str.replace("-1_", "_", regex=False)


# 6. Merge metadata with ATAC observations
merged_with_mapping = merged_obs.merge(mapping_fixed, on="cell_id", how="left")


# 7. Write merged metadata back to AnnData.obs
subset_mergeed.obs = merged_with_mapping.set_index(subset_mergeed.obs.index)


# 8. Clean column names
# Keep only annotation columns from the mapping table
subset_mergeed.obs = subset_mergeed.obs.rename(columns={
    "Class_label_name_y": "Class_label_name",
    "Subclass_label_name_y": "Subclass_label_name",
    "Group_label_name_y": "Group_label_name"
})

subset_mergeed.obs = subset_mergeed.obs.loc[
    :, ~subset_mergeed.obs.columns.str.endswith("_x")
]
```

```python
import pandas as pd

# ---------- 1) Construct a matching key from flitered_cell ----------
# The goal is to generate a cell_id consistent with merged_obs["cell_id"]

# Example flitered_cell index:
# AAACAGCCAAATACCT-1_D1-CaB-A
flc = flitered_cell.copy()

# Reset index to expose RNA cell IDs
flc = flc.reset_index().rename(columns={"index": "rna_cell_id"})

# Remove "-1_" to normalize format
# rna_cell_id: AAACAGCCAAATACCT-1_D1-CaB-A
# cell_id:     AAACAGCCAAATACCT_D1-CaB-A
flc["cell_id"] = flc["rna_cell_id"].str.replace("-1_", "_", regex=False)


# Columns to transfer from RNA metadata
# (modify as needed depending on required annotations)
cols_to_take = [
    "cell_id",
    "group", "sample", "donor", "target",
    "Neighborhood_label", "Neighborhood_name",
    "Class_label", "Class_name",
    "Subclass_label", "Subclass_name",
    "Group_label", "Group_name",
    "Group_bootstrapping_probability",
]

rna_anno = flc[cols_to_take].copy()


# ---------- 2) Merge RNA annotations into ATAC metadata ----------
# merged_obs already contains:
# atac_barcode, sample, rna_barcode, cell_id, etc.
# cell_id format: AAACAGCCAAATACCT_D1-CaB-A
merged_with_rna = merged_obs.merge(
    rna_anno,
    on="cell_id",
    how="left"
)


# ---------- 3) Write merged annotations back to subset_mergeed.obs ----------
subset_mergeed.obs = merged_with_rna.set_index(subset_mergeed.obs.index)


print(subset_mergeed.obs[["Subclass_name", "Group_name"]].head())
print(
    "Number of ATAC cells with matched RNA annotations:",
    subset_mergeed.obs["Subclass_name"].notna().sum()
)
```

```python
# Extract target label from sample_x
# Example: D1-CaB-A  →  target = A
subset_mergeed.obs["target"] = (
    subset_mergeed.obs["sample_x"]
    .str.extract(r'-([A-Z])(?:-\d+)?$')
)


# Extract donor ID from sample_x
# Example: D1-CaB-A → donor = D1
subset_mergeed.obs["donor"] = (
    subset_mergeed.obs["sample_x"]
    .str.extract(r'^([DM]\d+)')
)


# Function to extract region name from sample string
# Example:
#   D1-CaB-A  →  CaB
#   M2-SubTH-C → SubTH
import re

def extract_region(sample):
    match = re.search(r"-(\w+)-[A-C](-1)?$", sample)
    return match.group(1) if match else "Unknown"


# Apply region extraction to metadata
subset_mergeed.obs["region"] = (
    subset_mergeed.obs["sample_x"]
    .apply(extract_region)
)


# Quick check of parsed region labels
print(subset_mergeed.obs[["sample_x", "region"]].head())


# Normalize region naming for consistency
# (e.g., SubTH → STH)
subset_mergeed.obs["region"] = subset_mergeed.obs["region"].replace({
    "SubTH": "STH",
    "subTH": "STH"
})


# Final sanity check of extracted metadata
print(
    subset_mergeed.obs[["sample_x", "region", "target", "donor"]].head()
)
```
